```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr,annot=True,cmap='hot')
```

```python
plt.figure(figsize = (10,5))
sns.countplot('Region',hue='Category',data=df)
plt.title('Count-Region')
```

```python
plt.figure(figsize = (15,10))
sns.countplot('Region',hue='Sub-Category',data=df)
plt.title('Sub-Category/Region')
```

```python
sns.countplot(df['Ship Mode'])
```

```python
plt.figure(figsize = (15,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%.1f%%')
```

```python
df.groupby('Ship Mode')['Sales','Profit'].agg(['sum']).plot.bar()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
sns.pairplot(df,hue='Sub-Category')
```
